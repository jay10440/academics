#Some brief Data Analysis

```r
#Checking Correlations
cor(df)
```

```r
#Doing Pairwise Graphs
pairs(df[, c("mpg","disp","hp","wt","qsec")])
```

```r
#Shaded Area = Confidence Interval Band
ggplot(df, aes(x=wt, y=mpg))+geom_point()+geom_smooth(method="lm")+theme_minimal()
#Obviously less weight, more mpg, up to some deviation of course
```

```r
par(mfrow=c(2,2)) #Apparently does not work for ggplots...trying grid extra

plot1<-ggplot(df, aes(x=wt, y=mpg))+geom_point()+geom_smooth(method="lm")+theme_minimal()
plot2<-ggplot(df, aes(x=disp, y=mpg))+geom_point()+geom_smooth(method="lm")+theme_minimal()
plot3<-ggplot(df, aes(x=hp, y=mpg))+geom_point()+geom_smooth(method="lm")+theme_minimal()
plot4<- ggplot(df, aes(x=qsec, y=mpg))+geom_point()+geom_smooth(method="lm")+theme_minimal()
grid.arrange(plot1, plot2, plot3, plot4, nrow=2, ncol=2)
```

#Creating a Model and Making Predictions

None of these variables in particular seem to have anything suggestive of other fits. Linear models could work best.

```r
linmod_mtcars<- lm(mpg~ wt + hp + qsec, data=df)
linmod_mtcars
summary(linmod_mtcars)
```
